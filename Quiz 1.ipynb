```python
import numpy as np
from scipy.stats import kurtosis, skew
fuel_data1 = fuel_data['fuel_qty_burned']
print("Skew:")
skewValue = fuel_data1.skew()
kurtosisValue = fuel_data1.kurtosis()

print(skewValue)
print(kurtosisValue)
```
